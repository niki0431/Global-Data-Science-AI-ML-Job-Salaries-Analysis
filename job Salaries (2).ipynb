```python
df.columns
```

```python
def capping(data,columns):
    for i in columns:
        q1=data[i].quantile(0.25)
        q3=data[i].quantile(0.75)
        iqr=q3-q1
        lower=q1-(1.5*iqr)
        upper=q3+(1.5*iqr)
        data[i]=np.clip(data[i],lower,upper)
    
```

```python
nums_cols=df.select_dtypes([int,float])
nums_cols.columns
```

# feature engineering

```python
df.head(5)
```

```python
# Avoid division by zero
df['salary_per_remote_ratio'] = df['salary_in_usd'] / (df['remote_ratio'] + 1)
df
```

```python
df['is_same_country'] = (df['employee_residence'] == df['company_location']).astype(int)
df
```

```python
df['salary_per_year'] = df['salary_in_usd'] / (2025 - df['work_year'] + 1)
df
```

```python
df['salary_range'] = pd.qcut(df['salary_in_usd'], 4, labels=['Low', 'Med', 'High', 'V.High'])
df.head(5)
```
